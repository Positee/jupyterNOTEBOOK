```python
import sys
import matplotlib.pyplot as plt
import numpy as np

x=[1,2,3,4,5]
y=[2,4,6,8,10]

xp=np.array(x)
yp=np.array(y)

plt.plot(xp,yp)
plt.show()

plt.savefig(sys.stdout.buffer)
sys.stdout.flush()
```
